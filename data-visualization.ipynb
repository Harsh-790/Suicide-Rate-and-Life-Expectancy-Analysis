```python
dataset=pd.read_csv("/kaggle/input/suicide-rate-and-life-expectancy/Life expectancy.csv")
data=pd.read_csv("/kaggle/input/suicide-rate-and-life-expectancy/Suicide Rate.csv")
```

```python
country=[]
country1=data['Country'].values
country2=dataset['Country'].values
for i in country1:
    if i in country2:
        country.append(i)
```

```python
dataset=dataset.values
data=data.values
```

```python
Happy=[]#Happiness Score
Birth=[]#Fertility Rate (births per woman)
Men=[]#Life Expectancy (years) - Men
Women=[]#Life Expectancy (years) - Women
SR=[]#Suicide rate
GDP=[]#GDP per capita
for i in range(len(data)):
    if data[i][0] in country:
        Men.append(dataset[i][1])
        Women.append(dataset[i][2])
        Happy.append(dataset[i][3])
        Birth.append(dataset[i][4])
        SR.append(data[i][2])
        GDP.append(data[i][1])
for i in range(len(GDP)):
    GDP[i]=int(GDP[i][:len(GDP[i])-4]+GDP[i][len(GDP[i])-3:])
```

```python
plt.rcParams[ 'font.sans-serif' ] = [ 'SimHei' ]
plt.rcParams[ 'axes.unicode_minus' ] = False
plt.figure(figsize=(12,6))
plt.subplot(231);plt.xlabel("Happiness Score");plt.ylabel("Suicide rate");plt.scatter(Happy,SR)
plt.subplot(232);plt.xlabel("GDP per capita");plt.ylabel("Suicide rate");plt.scatter(GDP,SR)
plt.subplot(233);plt.xlabel("Fertility Rate (births per woman)");plt.ylabel("Suicide rate");plt.scatter(Birth,SR)
plt.subplot(234);plt.xlabel("Life Expectancy (years) - Men");plt.ylabel("Suicide rate");plt.scatter(Men,SR)
plt.subplot(235);plt.xlabel("Life Expectancy (years) - Women");plt.ylabel("Suicide rate");plt.scatter(Women,SR)
```
